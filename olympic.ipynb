```python
#Top 10 participates
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#Top 10 plot

plt.figure(figsize=(12,6))
plt.title('Overall Participants by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2');
```

```python
#Age Distribution
plt.figure(figsize=(12,6))
plt.title('Age Distribution of the Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color='#ADD8E6', edgecolor = 'Black');
```

```python
#Male and Female 
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
#pie cahrt male vs female
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels= gender_counts.index, autopct='%1.1f%%',startangle=100,shadow=True);
```

```python
#total medals
athletes_df.Medal.value_counts()
```

```python
#Total female athletes by olymic version
Female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
Female_participants = Female_participants.groupby('Year').count().reset_index()
Female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex=='F')& (athletes_df.Season=='Summer')]
```

```python
#Line graph Women Participants
part = womenOlympics.groupby('Year')['Sex'].value_counts()

plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes Over Time')
```

```python
#Gold Medal
GoldMedals =athletes_df[(athletes_df.Medal== 'Gold')]
GoldMedals
```

```python
#values different from NaN
GoldMedals = GoldMedals[np.isfinite(GoldMedals['Age'])]
GoldMedals
```

```python
#Gold mdedal over 55
GoldMedals['ID'][GoldMedals['Age'] > 55].count()
```

```python
sport_event =GoldMedals['Sport'][GoldMedals['Age']>55]
sport_event
```

```python
#Gold medal from each country
GoldMedals.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalgoldmedals = GoldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g= sns.catplot(x="index", y= "Medal", data =totalgoldmedals, height=5, kind ="bar", palette="rocket")
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Gold medals per countery")
plt.title('Gold Medals per Country')
```

```python
#Rio Olympics
max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year)&(athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y= team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('Countrywise Medals for the Year 2016')
```
